```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv("C://Users//apara//Documents//Python//UNZIP_FOR_NOTEBOOKS_FINAL (1)//05-Seaborn//dm_office_sales.csv")
```

<h2 style="color: red;"><u>Scatterplots with Seaborn</u></h2>

```python
plt.figure(figsize=(8, 3), dpi=200)
sns.scatterplot(data=df,x='salary',y='sales');
```

```python
fig, ax = plt.subplots(figsize=(8,3),dpi=200)
sns.scatterplot(
    data=df,
    x="sales",
    y="salary",
    hue="work experience",
    size='salary',
    ax=ax,
    palette='tab10'
)
ax.legend(
    bbox_to_anchor=(1.05, 1), 
    loc='upper left'
)
plt.show()
```

<b style="color: yellow;">For more information of the colormap/palette we can refer to this following documentation [link](https://matplotlib.org/stable/users/explain/colors/colormaps.html)</b>

```python
fig,ax=plt.subplots(figsize=(8,5),dpi=200)
sns.scatterplot(data=df,x='salary',y='sales',style='division',s=55,alpha=.7)
plt.legend(bbox_to_anchor=(1.05,1))
plt.show()
```

<h2 style="color: red"><b><u>Distribution Plots</u></b></h2>
Distribution plots display a single continuous feature and help visualize properties such as deviation and average values.

There are three main distribution plot types
1. Rug Plot
2. Histogram
3. KDE(Kernel Distribution Estimation)

```python
sns.rugplot(data=df,x='salary',height=.5)
```

```python
sns.displot(data=df,x='salary',kde=True,color='red');
```

```python
sns.kdeplot(data=df,x='salary')
plt.show()
```

<h2 style="color: red"><b><u>Categorical Plots</u></b></h2>

```python
df.head()
```

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10,5),dpi=200)
ax=sns.countplot(data=df,x='division',hue='division')

for container in ax.containers:
    ax.bar_label(
        container,
        label_type='center'   
    )
```

```python
df['level of education'].value_counts()
```

```python
plt.figure(figsize=(10,4),dpi=200)
ax=sns.countplot(data=df,x='level of education',hue='level of education',order=df['level of education'].value_counts().index)


for container in ax.containers:
    ax.bar_label(container, label_type='edge')
```

```python
plt.figure(figsize=(12,4),dpi=300)
ax = sns.countplot(data=df,x='level of education',hue='division',order=df['level of education'].value_counts().index,)

for container in ax.containers:
    ax.bar_label(
        container,
        label_type='edge'
    )
```

```python
plt.figure(figsize=(12,4), dpi=200)

ax = sns.barplot(
    data=df,
    x='level of education',
    y='salary',
    hue='division',
    estimator=np.mean,
    errorbar='sd'
)

for container in ax.containers:
    ax.bar_label(container, label_type='center',fmt='%.0f', fontsize=8)

plt.title("Average Salary by Education Level and Division", fontweight='bold')
plt.ylabel("Average Salary")
plt.tight_layout()
plt.legend(bbox_to_anchor=(1.05,1))
plt.show()
```

<h3 style="color: red;""><u>Categorical plots and distributions within the categories</u></h3>

Distribution within categories can be visualised by the following plots
1. Box Plot
2. Violin Plot
3. Swarm Plot
4. Boxen Plot(Lettr-Value Plot)

```python
plt.figure(figsize=(12,7),dpi= 200)
ax=sns.boxplot(data=df,x='level of education',y='salary',hue='level of education',palette='Set2',width=0.5)
```

```python
plt.figure(figsize=(12,7),dpi= 200)
ax=sns.violinplot(data=df,x='level of education',y='salary',hue='level of education',palette='Dark2',width=1)
```

```python
plt.figure(figsize=(12,7),dpi= 200)
ax=sns.swarmplot(data=df,x='division',y='salary',hue='division',palette='Dark2')
```

<h2 style="color: red;"><u>Comparison Plots</u></h2>

Comparison plots are essentially 2-dimensional versions of plot that we have learned so far.

The two main plot types are 
1. Jointplot
2. Pairplot

```python
sns.jointplot(data=df1,x='math score',y='reading score')
```

```python
sns.jointplot(data=df1,x='math score',y='reading score',hue='gender')
```

```python
sns.jointplot(data=df1,x='math score',y='reading score',kind='hex')
```

```python
sns.pairplot(data=df1,hue='gender',corner=True)
```
