<h1><b><u>Introduction to Matplotlib</b></u></h1>

* Matplotlib is known as "Grandfather" of plotting ans visualization libraries for Python
* Many other visualization libraries are built directly off Matpllotlib(e.g. seaborn and pandas built-in visulization)
* Matplotlib is heavily inspired by the plotting functions of MatLab programming language.
* It allows for the creation of almost any plot type and heavy customization.
* <span style="color: yellow;">There are actually two seperate approaches to creating plots</span>
   1. <span style="color: red;">Functional based methods</span>
   2. <span style="color: red;">OOP based method</span>

<h2 style="color: red;"><u>Functional based methods</u></h2>

The most basic way to use matplotlib is through the function plot callls:  
    plt.plot(x,y)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,11)
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y)
plt.xlabel('X Axis') # For X-axis title
plt.ylabel('Y Axis') # For Y-axis title
plt.title('String Title') # For plot title
plt.xlim(0,10) # Limits the X-axis within the given range i.e(0,10)
plt.ylim(0,20) # Limits the Y-axis within the given range i.e(0,20)
plt.savefig('1.myfirstplot.png');
```

<h2 style="color: red;"><u>OOP based method</u></h2>

```python
fig=plt.figure() # Like adding a blank canvas
axes=fig.add_axes([0,0,1,1]) # Adding axes to  the blank canvas
axes.plot(x,y); # Plotting the graph
```

```python
a=np.linspace(0,10,11)
a
```

```python
b=a**4
b
```

```python
x=np.arange(0,10)
y=2*x
print(x)
print(y)
```

```python
fig=plt.figure()
# Large axes
axes2=fig.add_axes([0,0,1,1])
axes2.plot(a,b)
axes2.set_xlim(0,8)
axes2.set_ylim(0,8000)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Power of 4')
# Small axes\
axes3=fig.add_axes([0.1,0.4,.5,.5])
axes3.plot(a,b)
axes3.set_xlim(1,2)
axes3.set_ylim(0,20)
axes3.set_xlabel('x')
axes3.set_ylabel('y');
axes3.set_title('Zoomed in');

fig.savefig('2.new_fig.png',bbox_inches='tight')
```

<h3 style="color: red;"><u>Subplots</u></h3>

```python
a=np.linspace(0,10,11)
b=a**4
```

```python
x=np.arange(0,10)
y=2*x
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2)
ax[0].plot(x,y)
ax[1].plot(a,b);
plt.tight_layout()
fig.subplots_adjust(wspace=1)
```

```python
x=np.linspace(-1,1,400)
y=np.sqrt(1-x**2)
```

```python
x1=np.linspace(-3,3,400)
y1=5*np.sqrt(1-(x1**2/9))
```

```python
x2=np.linspace(0,10,400)
y2=np.sqrt(x2**2-1)
```

```python
x3=np.linspace(0,10,400)
y3=np.sqrt(4*x3)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2,dpi=300,figsize=(12,10))

# ---- Circle ----
axes[0][0].plot(x, y,color='green')
axes[0][0].plot(x, -y,color='green')
axes[0][0].set_ylim(-1.2, 1.2)
axes[0][0].set_aspect('equal')     
axes[0][0].set_title("Circle")

# ---- Ellipse ----
axes[0][1].plot(x1, y1,color='purple')
axes[0][1].plot(x1, -y1,color='purple')
axes[0][1].set_ylim(-5.5, 5.5)
axes[0][1].set_aspect('equal')     
axes[0][1].set_title("Ellipse")

# ---- Hyperbola ----
axes[1][0].plot(x2, y2,color='red')
axes[1][0].plot(x2, -y2,color='red')
axes[1][0].plot(-x2, -y2,color='red')
axes[1][0].plot(-x2, y2,color='red')
axes[1][0].set_xlim(-10, 10)
axes[1][0].set_aspect('equal')     
axes[1][0].set_title("Hyperbola")

# ---- Parabola ----
axes[1][1].plot(x3,y3,color='blue')
axes[1][1].plot(x3,-y3,color='blue')
axes[1][1].set_aspect('equal')
axes[1][1].set_title('Parabola')
fig.subplots_adjust(wspace=2,hspace=2)

# ----Function to centre the axes----
def center_axes(ax):
    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('center')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

    
fig.suptitle('Graphs of different geomtric figures',fontsize=16) # Figure level title

for ax in axes.flat:
    center_axes(ax)


plt.tight_layout();

fig.savefig('3.subplots.png')
```

<h3><u><b>Some more practice </u></b></h3>

```python
a1=np.linspace(-2*np.pi,2*np.pi,2000)
b1=np.sin(a1)
```

```python
a2=np.linspace(-2*np.pi,2*np.pi,2000)
b2=np.cos(a2)
```

```python
a3=np.linspace(-2*np.pi,2*np.pi,2000)
b3=np.tan(a3)
b3[np.abs(b3) > 10] = np.nan
```

```python
a4=np.linspace(-2*np.pi,2*np.pi,2000)
b4=1/np.tan(a4)
b4[np.abs(b4) > 10] = np.nan
```

```python
fig1,axes1=plt.subplots(nrows=2,ncols=2,dpi=100,figsize=(12,7))

# ---- sin graph ----
axes1[0][0].plot(a1,b1)
axes1[0][0].set_title('sin(x) graph')


# ---- cos graph ----
axes1[0,1].plot(a2,b2,color='purple')
axes1[0,1].set_title('cos(x) graph')


# ---- tan graph ----
axes1[1,0].plot(a3,b3,color='Green')
axes1[1,0].set_title('tan(x) graph')
axes1[1,0].grid()


# ----cot graph ----
axes1[1,1].plot(a4,b4,color ='red')
axes1[1,1].set_title('cot(x) graph')
axes1[1,1].grid()



def cen(ax):
    ax.spines['bottom'].set_position('center')
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')


for ax in axes1.flat:
    cen(ax)

fig1.suptitle('Trigonometric Graph',fontsize=20)

plt.tight_layout();

fig1.savefig('4.trig.png')
```

<h3 style="color: red;"><u> Matplotlib Styling - Legends</u></h3>

<h3 style="color: red;"><u> Matplotlib Styling - Colors and Styles</u></h3>

```python
x=np.linspace(0,11,10)
y=x
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color='#7a2c05',lw=2)
axes.plot(x,y+1,color='#4d026e',lw=1,marker='o')
axes.plot(x,y+2,color='#e64f09',lw=1,marker='+',ms=15)
axes.plot(x,y+3,color='#126306',lw=3,marker='s',ms=10)
axes.plot(x,y+4,color='#126306',lw=3,linestyle='-.',marker='*',ms=10)
axes.plot(x,y+5,color='#e30bdc',lw=4,linestyle='-',marker='s',ms=10,markerfacecolor='#a8c906',markeredgewidth=2,markeredgecolor='#8a0404')

# ---- Customizing linestyle ----
lines=axes.plot(x,y+6,color='#b00505',lw=3,marker='h',ms=10)
lines[0].set_dashes([1,2,1,2,5,2]) # A line with 1 lenghth-2space- 1 length-2space-5length-2space
```
